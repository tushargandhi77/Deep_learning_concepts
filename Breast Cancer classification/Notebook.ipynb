```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
```

```python
def dianosis(text):
    if text == 'M':
        return 1
    else:
        return 0
```

```python
df['diagnosis'] = df['diagnosis'].apply(dianosis)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
df.isna().sum()
```

```python
from sklearn.decomposition import PCA
```

```python
scaled_df = df.drop(columns=['diagnosis'])

for i in scaled_df.columns:
    scaled_df[i] = (scaled_df[i] - scaled_df[i].mean())/scaled_df[i].std()
```

```python
pca = PCA()
pca = pca.fit(scaled_df)

display(pca.explained_variance_ratio_)
```

```python
pca_fit = PCA(n_components=2).fit_transform(scaled_df)

pcs_df = pd.DataFrame(data=pca_fit,columns=['PC1','PC2'])
pcs_df['diagnosis'] = df.diagnosis.values
```

```python
pcs_df
```

```python
sns.scatterplot(x=pcs_df['PC1'],y=pcs_df['PC2'],hue=pcs_df['diagnosis'])
plt.title('PCA Graph')
plt.xlabel('PC1')
plt.ylabel('PC2')
```

```python
scaled_df
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(1,activation='sigmoid',input_dim=30))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics = ['accuracy'])
```

```python
history = model.fit(scaled_df,df['diagnosis'],epochs = 100,validation_split=0.2)
```

```python
model.layers[0].get_weights()
```

```python
model.layers[1].get_weights()
```

```python
plt.plot(history.history['loss'])
```

```python
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
```

```python
y_log = model.predict(scaled_df)
```

```python
y_pred = np.where(y_log>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(df['diagnosis'],y_pred)
```
