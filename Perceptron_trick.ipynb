```python
from sklearn.datasets import make_classification 
import numpy as np
X,y = make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=31,hypercube=False,class_sep=1)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
def perceptron(X,y):
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(np.shape(X[1]))
    lr = 0.1
    for i in range(1000):
        j =np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr*(y[j] - y_hat)*X[j]
    return weights[0],weights[1:]
```

```python
_intercept,_coef = perceptron(X,y)
```

```python
print(_intercept)
print(_coef)
```

```python
m = -(_coef[0]/_coef[1])
b = -(_intercept/_coef[1])
```

```python
X_input = np.linspace(-3,3,100)
y_input = m*X_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(X_input,y_input,color='red')
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
